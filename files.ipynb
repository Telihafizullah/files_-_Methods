```python
# Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode
# of file opening.
#  The function used to open a file is called "open()" and it takes the filename and the mode as arguments; the different modes for opening 
#  a file include: "r" for reading, "w" for writing, "a" for appending, and combinations like "r+" for both reading and writing to a file

# Q2. Why close() function is used? Why is it important to close a file?
# The close() function is used to explicitly tell the operating system that a file is no longer needed, releasing the system resources 
# associated with that open file; this is crucial because failing to close a file can lead to data loss, corruption, and potential issues 
# with exceeding the system's limit on open files simultaneously

# Q3. Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
# close the file. Open this file and read the content of the file.
# Create and write to the file
with open("data_scientist.txt", "w") as file: 
    file.write("I want to become a Data Scientist") 

# Open and read the file
with open("data_scientist.txt", "r") as file: 
    content = file.read()
    print(content)

# Q4. Explain the following with python code: read(), readline() and readlines().
# read():
# Reads the entire contents of a file as a single string.
# Useful when you want to process the whole file at once
with open("myfile.txt", "r") as file:
    content = file.read()
    print(content)

# readline():
# Reads a single line from a file.
# Returns the line as a string, including the newline character (\n).
# Useful when you want to process the file line by line.
with open("myfile.txt", "r") as file:
    line = file.readline()
    while line:
        print(line, end="")  # Use end="" to prevent extra newlines
        line = file.readline()

# Reads all lines from a file and returns them as a list of strings.
# Each element of the list represents a line from the file.
# Useful when you want to work with the lines as individual elements. 
 
with open("myfile.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        print(line, end="")


# Q5. Explain why with statement is used with open(). What is the advantage of using with statement and
# open() together?
# The "with" statement is used with the "open()" function in Python to automatically close a file once you are finished using it
# , eliminating the need to explicitly call the "close()" method, thus preventing potential errors that could occur if you forget to
# close the file manually, especially in situations where exceptions might be raised; this makes your code cleaner and more reliable.

# Q6. Explain the write() and writelines() functions. Give a suitable example.
# The `Write` method is used to write text to the console without adding a newline character at the end of the text. 
# The `WriteLine` method is used to write text to the console and adds a newline character at the end of the text, causing the next 
# output to appear on a new line
```
